```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob 
import os


#adding d=0 as a counter for saving the image in a different names
d=0

#adding the source path of the first section of the dataset
SOURCEPATH = glob.glob("I:\\UNI FILES\\3rd sem\\Image Processing and Computer Vision\\Final sem 2\\leaf_dataset\\leaves_testing_set_1\\INPUTS\\*.png")

    
#adding the path of the first output folder    
OUTPUTPATH = ("I:\\UNI FILES\\3rd sem\\Image Processing and Computer Vision\\Final sem 2\\leaf_dataset\\leaves_testing_set_1\\OUTPUTS\\")

#making sure the path exists

if not os.path.exists(OUTPUTPATH):
    print("path "  + SOURCEPATH + " DOES NOT EXIST!")
    exit(1)
#for loop to go through the whole images in the dataset    
for file in SOURCEPATH:
    
    img = cv2.imread(file)
    
    if not os.path.exists(file):
        print("path "  + SOURCEPATH + " DOES NOT EXIST!")
        exit(1)
    
    
#converting to HSV colorspace 
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    plt.imshow(hsv)
    plt.show()

    cv2.imshow('image mask', hsv)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


#reducing the noise, smoothenng the image
    img_s_blur = cv2.GaussianBlur(hsv, (7, 7), 0)  
    plt.imshow(img_s_blur)
    plt.show()

    cv2.imshow('image mask2', img_s_blur)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


#choosing the correct HSV boundaries to segment out the leaf 

    mask = cv2.inRange(img_s_blur, (25, 60, 5), (70, 255,255))
    plt.imshow(mask)
    plt.show()

    cv2.imshow('image mask2', mask)
    cv2.waitKey(0)
    cv2.destroyAllWindows()




#saving the image to the output folder
    cv2.imwrite(OUTPUTPATH + str(d) + ".png", mask)
    d+=1
    

#repeating for the second folder  
d=0
SOURCEPATH = glob.glob("I:\\UNI FILES\\3rd sem\\Image Processing and Computer Vision\\Final sem 2\\leaf_dataset\\leaves_testing_set_2\\color_images\\*.png")
OUTPUTPATH = ("I:\\UNI FILES\\3rd sem\\Image Processing and Computer Vision\\Final sem 2\\leaf_dataset\\leaves_testing_set_1\\OUTPUTS2\\")
if not os.path.exists(OUTPUTPATH):
    print("path "  + SOURCEPATH + " DOES NOT EXIST!")
    exit(1)
    
for file in SOURCEPATH:
    
    img = cv2.imread(file)
    
    if not os.path.exists(file):
        print("path "  + SOURCEPATH + " DOES NOT EXIST!")
        exit(1)

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img_s_blur = cv2.GaussianBlur(hsv, (7, 7), 0)  
    mask = cv2.inRange(img_s_blur, (25, 60, 5), (70, 255,255))






    cv2.imwrite(OUTPUTPATH + str(d) + ".png", mask)
    d+=1
```
